1. What is the difference between a function and a method in Python?
  - A function is independent, while a method is associated with an object.
2. Explain the concept of function arguments and parameters in Python.
  - Parameters are variables in function definitions while arguments are actual values passed.
  - Example: def add(a, b): return a + b (here a, b are parameters), add(2, 3) (here 2,3 are arguments).
3.  What are the different ways to define and call a function in Python?
  - The function is defined as "def func_name(): pass".
  - The function is called as "func_name(), func_name(arg)".
4. What is the purpose of the `return` statement in a Python function?
  - Returns a value from a function and exits execution.
  - Example:
  - def square(x): return x*x
5.  What are iterators in Python and how do they differ from iterables?
  - Iterators is an object that keeps track of the iteration, providing the next item using `__next__()` and raising `StopIteration` when done. While Iterables is an object that can return an iterator using `__iter()__`.
6. Explain the concept of generators in Python and how they are defined.
  - A generator in Python is a special type of iterator that generates items one at a time using the yield keyword. Unlike regular functions, which return a value and exit, a generator yields values and can resume execution from where it left off.
7. What are the advantages of using generators over regular functions?
  - The advantages of using generators over regular functions are better memory efficient, and faster execution for large datasets.
8. What is a lambda function in Python and when is it typically used?
  - A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but can only have a single expression.
  - It is used when we need a function for a short task, without formally defining a full function using `def`.
9. Explain the purpose and usage of the `map()` function in Python?
  - The map() function in Python applies a given function to all items in an iterable such as lists, and returns a map object.
  - map() is used to modify each element in an iterable using a function in a readable way than using a for loop.
10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - `map()` applies a function to all elements. Example: map(lambda x: x*2, [1,2,3]) -> [2,4,6].
  - `reduce()` aggregates elements into a single value. Example: reduce(lambda x,y: x+y, [1,2,3]) -> 6.
  - `filter()` selects elements based on condition. Example: filter(lambda x: x%2==0, [1,2,3]) -> [2].

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even(lst):
    return sum(x for x in lst if x % 2 == 0)
input_str = input("Enter numbers separated by space: ")
lst = list(map(int, input_str.split()))
print(sum_even(lst))
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(s):
    return s[::-1]
st = input("Enter a string: ")
print(reverse_string(st))
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number
def square_list(lst):
    return [x**2 for x in lst]
input_str = input("Enter numbers separated by space: ")
lst = list(map(int, input_str.split()))
new_list = square_list(lst)
print(new_list)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for x in range(2,n):
      if(n%x==0):
        return False
    return True
num = int(input("Enter any number between 1 and 200: "))
if(is_prime(num)):
  print(f"{num} is a prime number")
else:
  print(f"{num} is not a prime number")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n, self.a, self.b = n, 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.n <= 0:
            raise StopIteration
        self.n -= 1
        self.a, self.b = self.b, self.a + self.b
        return self.a
fib_seq = Fibonacci(10)
for num in fib_seq:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(n):
    for i in range(n+1):
        yield 2**i
gen = power_of_2(7)
for num in gen:
    print(num, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def file_line_generator(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()
file_name = 'sonnet.txt'
for line in file_line_generator(file_name):
    print(line)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
def sorting(tuples_list):
    return sorted(tuples_list, key=lambda x: x[1])
T = [(1, 3), (2, 1), (4, 2), (5, 0)]
answer = sorting(T)
print(answer)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def cTof(C):
    return (C * 9/5) + 32
def convert(C):
    return list(map(cTof, C))
C = [0, 20, 37, 100]
F = convert(C)
print(F)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(str):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', str))
str = input("Enter a string: ")
answer = remove_vowels(str)
print(answer)
```

```python
#11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
"""Order Number             Book Title and Author             Quantity            Price per Item
   34587                    Learning Python, Mark Lutz            4                    40.95
   98768                    Programming Python, Mark Lutz         5                    56.80
   77226                    Head First Python, Paul Berry         3                    32.95
   88112                    Einfuhrung in Python3, Bernd Klein    3                    24.99
"""
# Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
# product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
# order is smaller than 100,00€.
# Write a Python program using lambda and map

orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98768, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Berry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

def calculate_order_value(order):
    order_number, _, quantity, price = order
    total = quantity * price
    if total < 100:
        total += 10
    return (order_number, total)

total_orders = list(map(lambda order: calculate_order_value(order), orders))
print(total_orders)
```
